```python
import cv2
import glob
import numpy as np

from tqdm import tqdm
```

```python
TRAIN_DATA = glob.glob("dataset/train/mask/*")
VAL_DATA = glob.glob("dataset/val/mask/*")
TEST_DATA = glob.glob("dataset/test/mask/*")
```

```python
unique_channels = []
```

```python
with tqdm(total=len(TRAIN_DATA)) as pbar:
    for i, filepath in enumerate(TRAIN_DATA):
        pbar.set_description('Extracting: %d' % (1 + i))
        pbar.update(1)
        # Load an image
        image = cv2.imread(filepath)
        # Get the number of channels in the image
        num_channels = image.shape[2]
        # Loop over each channel and get its unique values
        for i in range(num_channels):
            channel = image[:, :, i]
            unique_values = np.unique(channel)
            unique_channels = unique_channels + list(unique_values)
            unique_channels = list(np.unique(np.asarray(unique_channels)))
```
